```python
import math
import matplotlib.pyplot as plt
m = 68.1
g = 9.8
c = 12.5
def f(t, v):
    return(m*g-c*v)/m
def fa(t, v):
    return(g*m/c)*(1-math.exp(-1*c*t/m))
v = 0
va = 0
vm = 0
t = 0
h = 1
tar = []
var = []
vanar = []
vmar = []
for i in range(0, 10):
    t = t+h
    v = v + f(t, v)
    vm = vm + f(t+h/2, vm+f(t, vm)*h/2)*h
    va = fa(t, va)
    tar.append(t)
    var.append(v)
    vanar.append(va)
    vmar.append(vm)
    print("t", i, " = ", t, " v", i, " = ", v," midpoint v = ", vm, " analytical v = ", va)
plt.plot(tar, var)
plt.plot(tar, vanar)
plt.plot(tar, vmar)
plt.legend(['Euler', 'Analytical', 'Midpoint'])
```
