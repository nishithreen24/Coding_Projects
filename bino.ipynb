```python
import numpy as np
import matplotlib.pyplot as  plt
import math
def binomial_pdf(k, n, p):
    return (math.comb(n, k))*(p**k)*((1-p)**(n-k))
N = [10, 50, 100, 500]
x = np.random.binomial(500, 0.5, 500)
xv = np.arange(x.min(), x.max()+1)
yv = [binomial_pdf(k, 500, 0.5) for k in xv]
bins=np.arange(x.min(), x.max()+1)
plt.plot(xv, yv, color='red')
plt.title("binomial probability distribution function")
plt.show()
```
