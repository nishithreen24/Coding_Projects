```python
import numpy as np
import matplotlib.pyplot as plt
X = np.arange(-100, 100, 0.001)
m = -1
c = 1
Y = (m*X)+c
plt.subplot(1, 2, 1)
plt.grid()
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title("Z-Plane")
plt.plot(X, Y)
U = X/(X**2+Y**2)
V = -Y/(X**2+Y**2)
plt.subplot(1, 2, 2)
plt.grid()
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title("W-Plane")
plt.plot(U, V)
```
