```python
import matplotlib.pyplot as plt
import numpy as np
m = 1
c = 0
def func(x):
    return (m*x)+c
X = np.linspace(-10, 10, 1000)
Y = func(X)
plt.subplot(1, 2, 1)
plt.title("Z-Plane")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.plot(X, Y)
plt.subplot(1, 2, 2)
plt.title("W-Plane")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
U = X/((X**2)+(Y**2))
V = -Y/((X**2)+(Y**2))
plt.plot(U, V)
```
