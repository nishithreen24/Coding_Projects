```python
import numpy as np
import matplotlib.pyplot as plt
c=int(input("enter the number "))
L1=[]
L2=[]
L3=[]
L4=[]
for i in range(0,c,20):
        x=np.random.poisson(7,i)
        x_=np.mean(x)
        var1=np.var(x)
        q=abs(x_-var1)
        L1.append(i)
        L2.append(x_)
        L3.append(var1)
        L4.append(q)


w=np.array(L1)
s=np.array(L2)
p=np.array(L3)
r=np.array(L4)
plt.plot(w,s)
plt.plot(w,p)
plt.plot(w,r)
plt.xlabel('data')
plt.title('poission distribution')
plt.show()
```
