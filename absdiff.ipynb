```python
import numpy as np
import matplotlib.pyplot as plt
lam = eval(input("enter lambda: "))
mean = []
variance = []
abs_diff = []
for i in range(0, 1000, 10):
    n = np.random.poisson(lam, i)
    mean.append(np.mean(n))
    variance.append(np.var(n))
    abs_diff.append(abs(np.mean(n)-np.var(n)))
plt.plot(mean)
plt.plot(variance)
plt.plot(abs_diff)
plt.show
```
