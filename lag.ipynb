```python
import numpy as np
from scipy import special
def f(x):
    return(np.exp(x**2)/(1+x**6))
n = 100
xi = special.roots_hermite(n)[0]
wi = special.roots_hermite(n)[1]
print(xi)
#a = 0
#b = 2*np.pi
sum = 0
#c = (a+b)/2
#m = (b-a)/2
for i in range(0, n):
    sum = sum + wi[i]*f(xi[i])
print(sum/2)
```
