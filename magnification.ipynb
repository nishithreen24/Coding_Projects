```python
import numpy as np
import matplotlib.pyplot as plt
def plotm(m, c, s, e):
    if (m==np.tan((np.pi/2))):
        X = np.linspace(s, e, 100)
        Y = X*0+c
        plt.plot(Y, X)
    else:
        X = np.linspace(s, e, 100)
        Y = (X*m)+c
        plt.plot(X, Y)
sx = 0
sy = 0
ex = 1
ey = 2
m1 = 0
m2 = 0
m3 = np.tan((np.pi/2))
m4 = np.tan((np.pi/2))
c1 = 0
c2 = 2
c3 = 0
c4 = 1
plt.figure(figsize=(5, 5))
plt.grid()
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plotm(m1, c1, sx, ex)
plotm(m2, c2, sx, ex)
plotm(m3, c3, sy, ey)
plotm(m4, c4, sy, ey)
plt.show()
m = 2.5
plt.figure(figsize=(5, 5))
plt.grid()
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plotm(m1, c1*m, sx*m, ex*m)
plotm(m2, c2*m, sx*m, ex*m)
plotm(m3, c3*m, sy*m, ey*m)
plotm(m4, c4*m, sy*m, ey*m)
```
