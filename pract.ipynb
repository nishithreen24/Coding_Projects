```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return (x**3 + x**2 - 4*x + 1)
a = -2
b = 1
tol = 0.000001
while((b-a) >= tol):
    c = (b+a)/2
    if((f(b)*f(a))>0):
        print("No root between these limits")
        break
    elif(f(a)==0):
        print("root is ", a)
        break
    elif(f(b)==0):
        print("root is ", b)
        break
    elif(f(c)==0):
        print("root is ", c)
        break
    elif((f(a)*f(c))<0):
        print("closest value of root is ", c)
        b = c
    else:
        print("closest value of root is ", c)
        a = c
print("root is ", c)
x = np.arange(-11, 11, 0.01)
z = x*0
plt.plot(x, f(x))
plt.plot(z, x)
plt.plot(x, z)
plt.xlim(-5, 5)
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return(np.cos(x)-x)
def fd(x):
    return(-1*np.sin(x)-1)
def formula(x0):
    return(x0-(f(x0)/fd(x0)))
def newtonraphson(x0):
    tol = 0.0000001
    x1 = formula(x0)
    while((x1-x0)>tol):
        x0 = x1
        x1 = formula(x1)
    print(x1)
newtonraphson(2)
x = np.arange(-1, 1, 0.0001)
plt.plot(x, f(x))
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return(np.tan(x)-x)
def formula(x0, x1):
    return(((f(x1)*x0)-f(x0)*x1)/(f(x1)-f(x0)))
def secant(x0, x1):
    tol = 0.00001
    x2 = formula(x0, x1)
    while((x2-x1)>=tol):
        x0 = x1
        x1 = x2
        x2 = formula(x1, x2)
    print(x2)
secant(4.15, 4.55)
x = np.arange(0, 5, 0.001)
plt.plot(x, f(x))
plt.ylim(-1, 1)
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt
l = 0.05
t = 0
N0 = 1000
def f(t, N):
    return (-1*l*N)
def fa(t):
    return(N0*np.exp(-1*l*t))
N = N0
h = 0.5
T = []
a = []
ar = []
for i in range(1, 199):
    N = N + (f(t, N)+f(t+h, N+f(t, N)*h))*h/2
    T.append(t)
    a.append(N)
    ar.append(fa(t))
    t = t + h
print(t, " - ", N)
print(t, " - ", fa(t))
plt.plot(T, a)
plt.plot(T, ar)
plt.legend(['Heun', 'Analytical'])
plt.grid()
```
