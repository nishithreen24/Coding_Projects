```python
import math
import matplotlib.pyplot as plt
def f1(x, y):
    return(((x**3)*math.exp(-2*x))-(2*y))
def f2(x, y):
    return(x-3)
x = -1
y = 1.847
h = 0.1
X = []
Y = []
for i in range(0, 100):
    y = y + (f1(x, y) + f1(x+h, y+f1(x, y)*h))*h/2
    x = x + h
    Y.append(y)
    X.append(x)
    i = i + 1
    print("x", i, " = ", x, " y", i, " = ", y)
plt.plot(X, Y)
```
