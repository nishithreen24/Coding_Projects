```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 300*x/(1+np.exp(x))
N = range(1, 16)
E = []
for j in N:
    a = 0
    b = 10
    n = j
    h = (b-a)/n
    A = np.arange(a, b, h)
    F = []
    integral = 0
    for i in A:
        integral = integral + ((f(i)+f(i+h))*(h))/2
        F.append(f(i))
    print(integral)
    error = np.abs(246.59029-integral)*100/246.59029
    print(n, " - ", error)
    E.append(error)
B = np.arange(0, 10, 0.000001)
plt.plot(N, E)
plt.xlabel('number of subintervals')
plt.ylabel('error %')
plt.show()
plt.plot(B, f(B))
plt.ylabel('f(x)')

```
