```python
import numpy as np
import matplotlib.pyplot as plt
def f(t):
    return 2000*np.log(140000/(140000-2100*t))-9.8*t
N = range(1, 20)
E = []
for j in N:
    a = 8
    b = 30
    n = j
    h = (b-a)/n
    A = np.arange(a, b, h)
    F = []
    integral = 0
    for i in A:
        integral = integral + (h)*1/6*(f(i)+4*f((2*i+h)/2)+f(i+h))
        F.append(f(i))
    print(integral)
    error = 100*np.abs(11061-integral)/11061
    print(n, " - ", error)
    E.append(error)
B = np.arange(0, 10, 0.000001)
plt.plot(N, E)
plt.xlabel('number of subintervals')
plt.ylabel('error %')
plt.grid()
plt.show()
plt.plot(B, f(B))
plt.ylabel('f(x)')
plt.grid()
plt.show()
```
