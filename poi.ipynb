```python
import numpy as np
import matplotlib.pyplot as plt
import math

def poisson_pdf(k, lam):
    return (np.exp(-lam) * (lam**k)) / math.factorial(k)

def plot_poisson_pdf(lam):
    x = np.random.poisson(lam, 1000)
    x_v = np.arange(x.min(), x.max()+1)
    y_v = [poisson_pdf(k, lam) for k in x_v]
    #plt.hist(x, bins=np.arange(x.min(), x.max()+1), density=True, alpha=0.5)
    bins=np.arange(x.min(), x.max()+1)
    plt.plot(x_v, y_v, color='red', label='Poisson PDF')
    plt.xlabel('Number of Events (k)')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()
lam = 10
plot_poisson_pdf(lam)
```
