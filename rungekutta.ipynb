```python
import numpy as np
import matplotlib.pyplot as plt
l = 0.05
def f(t, N):
    return(-1*l*N)
def rungekutta(t0, N0, dt, n):
    t = np.zeros(n+1)
    N = np.zeros(n+1)
    t[0] = t0
    N[0] = N0
    for i in range(n):
        k1 = dt*f(t[i], N[i])
        k2 = dt*f(t[i]+dt/2, N[i]+k1/2)
        k3 = dt*f(t[i]+dt/2, N[i]+k2/2)
        k4 = dt*f(t[i]+dt, N[i]+k3)
        t[i+1] = t[i] + dt
        N[i+1] = N[i] + (k1+2*k2+2*k3+k4)/6
    return t, N
t0 = 0.0
N0 = 1000.0
dt = 1
n = 100
t, N = rungekutta(t0, N0, dt, n)
for i in range(n):
    print("t = ", t[i], "seconds    N = ", N[i])
plt.plot(t, N)
plt.ylabel('Number of nuclei remaining(N)')
plt.xlabel('Time(t)')
```
