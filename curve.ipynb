```python
import numpy as np
import matplotlib.pyplot as plt
def gplot(distribution, n):
    if distribution == "uniform":
        data = np.random.uniform(0, 1, n)
    elif distribution == "binomial":
        data = np.random.binomial(500, 0.5, n)
    elif distribution == "poisson":
        data = np.random.poisson(50, n)
    elif distribution == "gaussian":
        data = np.random.normal(0, 1, n)
    else:
        raise ValueError("Invalid distribution name")
    str = distribution+' distribution with N = ', n
    plt.plot(data)
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.title(str)
    plt.show()
N = [10, 50, 100, 500]
for distribution in ["uniform", "binomial", "poisson", "gaussian"]:
    for n in N:
        gplot(distribution, n)
```
