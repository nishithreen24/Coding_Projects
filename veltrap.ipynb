```python
import numpy as np
import matplotlib.pyplot as plt
def f(t):
    return 2000*np.log(140000/(140000-2100*t))-9.8*t
N = range(1, 11)
E = []
for j in N:
    a = 8
    b = 30
    n = j
    h = (b-a)/n
    A = np.arange(a, b, h)
    F = []
    integral = 0
    for i in A:
        integral = integral + ((f(i)+f(i+h))*(h))/2
        F.append(f(i))
    print(integral)
    error = np.abs(11061-integral)*100/11061
    print(n, " - ", error, "%")
    E.append(error)
plt.plot(N, E)
plt.xlabel('number of subintervals')
plt.ylabel('error %')
plt.show()
plt.plot(A, F)
plt.xlabel('time')
plt.ylabel('velocity')
```
