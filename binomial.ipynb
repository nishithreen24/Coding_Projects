```python
import numpy as np
import matplotlib.pyplot as plt
import math
def factorial(x):
    if(x>1):
        result = x*factorial(x-1)
    else:
        result = 1
    return result
n, p = 10, .5  # number of trials, probability of each trial
s = np.random.binomial(n, p, 1000)
count, x, ignored = plt.hist(s, 30, density=True)
plt.plot(x, factorial(n)/(factorial(x)*(factorial(n-x)))*p**x*(1-p)**(n-x), linewidth=2, color='r')
```
