```python
import math
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return (np.tan(x)-x)
def bisection(a, b, tol):
    if((f(a)*f(b))>0):
        print("Invalid Limits for this Function")
    else:
        while(((b-a)/2)>tol):
            c = (a+b)/2
            if(f(a)==0):
                print("The root of the Equation is ", a)
                break
            elif(f(b)==0):
                print("The root of the Equation is ", b)
                break
            elif(f(c)==0):
                print("The root of the Equation is ", c)
                break
            elif(f(a)*f(c)<0):
                b = c
            else:
                a = c
                print("The closest value of the root is ", c)
def plot():
    x = np.linspace(-11, 11, 10000000)
    y = np.tan(x)-x
    z = x*0
    plt.plot(x, y)
    plt.plot(x, z)
    plt.plot(z, x)
    plt.ylim(-10, 10)
    plt.xlim(-5, 5)
bisection(-4.5, -4, 0.00001)
plot()
```
