```python
#Numerical integration
#Simpson's method
import numpy as np
import matplotlib.pyplot as plt
def f(t):
    return 300*t/(1+np.exp(t))
a=eval(input("enter the initial value of interval :"))
b=eval(input("enter the final value of interval :"))
n=eval(input("enter the number of segament :"))
i=1
X=[]
Y=[]
while i<(n+1):
    h=abs(b-a)/i
    A=np.arange(a,b,h)
    I=0
    for j in A:
        I=I+(h)*1/6*(f(j)+4*f((2*j+h)/2)+f(j+h))
    print('integral for n=',i,I)
    i=i+1
    e = (abs(I - 246.59029) / 246.59029) * 100
    X.append(i)
    Y.append(e)
    print('for n=',i,'error','-',e)
    i = i + 1
x = np.linspace(0, n, 1000)
plt.subplot(2, 1, 1)
plt.plot(x, f(x))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.subplot(2, 1, 2)
plt.plot(X, Y)
plt.ylabel("error")
plt.xlabel("no. of segament")
plt.grid()
plt.show()
```
