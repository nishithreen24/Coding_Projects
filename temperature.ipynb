```python
import math
import matplotlib.pyplot as plt
k = 0.02
Ta = 20
T0 = 30
def f(t, T):
    return(-1*k*(T-Ta))
def fa(t):
    return((T0-Ta)*math.exp(-1*k*t))+Ta
t = 0
T = T0
Tmid = T0
midarray = []
anarray=[]
h = 1
tarray = []
Tarray = []
for i in range(0, 100):
    t = t + h
    T = T + f(t, T)*h
    Tmid = Tmid + f(t+h/2, Tmid+f(t, Tmid)*h/2)*h
    Tan = fa(t)
    Tarray.append(T)
    tarray.append(t)
    anarray.append(Tan)
    midarray.append(Tmid)
    i = i + 1
    print("t", i, " = ", t, " T", i, " = ", T," midpoint T = ", Tmid," analytical T = ", Tan)
plt.plot(tarray, Tarray)
plt.plot(tarray, anarray)
plt.plot(tarray, midarray)
plt.legend(['Euler', 'Analytical', 'Midpoint'])
```
